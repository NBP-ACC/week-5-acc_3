### a) Preprocessing the data [1 pt]

First of all you should preprocess the data. This is an important step to avoid running into any problems when you start to analyse the data. Since we provide the Seahaven dataset as an excel file make sure to run ```pip install xlrd``` in your activated acc environment beforehand. This allows you to directly read from excel files.

- Import the data of all three tasks (Absolute, Relative, Pointing) into one dataframe. Since we don't need the whole data, load only the columns "ReactionTime", "AngularDiffBin", "Subject", "Task", "Time", "Answer" and "Measurement".
- Clean the dataframe of all NaNs, i.e. remove all rows where at least one element is missing. How many rows have been removed?
- Change the values of the column "Answer". Replace each value "wrong" with 0 and each value "correct" with 1.

````python
xlsx = pd.ExcelFile(os.getcwd() + '/Seahaven_dataset.xlsx')
columns = ['ReactionTime', 'AngularDiffBin', 'Subject', 'Task', 'Time', 'Answer', 'Measurement']
df = pd.read_excel(xlsx, 'AbsoluteTask', usecols=columns)
df = pd.concat((df, pd.read_excel(xlsx, 'RelativeTask', usecols=columns)), axis=0)
df = pd.concat((df, pd.read_excel(xlsx, 'PointingTask', usecols=columns)), axis=0)
df = df.reset_index()
df = df.drop(labels='index', axis=1)

len_before_drop = len(df)
df = df.dropna()

df['Answer'] = df['Answer'].map({'wrong':0, 'correct':1})  # rename values in column 'Answer'

print('{} rows have been removed due to missing data\n'.format(len_before_drop - len(df)))
print(df.head())
````

You may have noticed that there are (extreme) outliers, that have to be removed from the **preprocessed data**.

- For each task (Absolute, Relative, Pointing) and time condition (3sec, Infinite) look at the RTs and keep only the ones that are within +2 and -2 standard deviation:

$ |(RT_{group1} - mean(RT_{group1}))| \leq (2*std(RT_{group1})) $

$group1$ = e.g. data of absolute task for time condition 3sec

- Make again a violinplot for each combination of task and time condition (you should end up with 6 violinplots).
````python
# Outlier removal: 727 trials are considered as outliers
stds = 2.0
outliers = df[['Task', 'Time', 'ReactionTime']].groupby(['Task','Time']).transform(
           lambda group: (group - group.mean()).abs().div(group.std())) > stds

df_preproc = df[outliers.ReactionTime == False]
df_preproc.reset_index()

print('{} trials are considered outliers'.format(len(df) - len(df_preproc)))

# df_preproc = df_preproc.reset_index()
# df_preproc = df_preproc.drop(labels='index', axis=1)
# distributions after outlier removel
means_preproc = df_preproc.groupby(['Subject', 'Task', 'Time'])['ReactionTime'].mean().reset_index()

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(8,6))
violins = sns.violinplot(data=means_preproc, x='Task', y='ReactionTime', hue='Time', ax=axes)
violins.set(ylabel='Reaction Time (s)', xlabel='Taks conditions', ylim=[-3, 28], title='Distributions after outlier removal')
fig.tight_layout()
fig.show()
````

### a) Hypothesis 1 [2 pt]

**Hypothesis 1:** "Given that subjects are limited by time, if they are faster in RT they are also less accurate."

- Use the data of the relative task.
- For each time condition (3sec / Infinite) calculate the mean RT and Accuracy per subject. Rename the column "Answer" to "Accuracy" (the accuracy corresponds to the averaged answer-values).
- Make a scatterplot of the mean RT (x-axis) and the accuracy (y-axis) for the time condition "3sec". 
- Make a second scatterplot and add a simple linear regression line to it. Calculate the slope of the regression line (you are allowed to use scipy). 
- Considering your results, what can you say about the hypothesis?

````python
rel = df_preproc[df_preproc['Task'] == 'Relative']
rt_acc = pd.DataFrame()
rt_acc['Mean RT'] = rel.groupby(['Subject', 'Time'])['ReactionTime'].mean()
rt_acc['Accuracy'] = rel.groupby(['Subject', 'Time'])['Answer'].mean()

rt_acc = rt_acc.reset_index()

rt = rt_acc[rt_acc['Time'] == '3sec']['Mean RT']
acc = rt_acc[rt_acc['Time'] == '3sec']['Accuracy']
slope, intercept, rvalue, pvalue, stderr = stats.linregress(rt, acc)  # linear regression
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html
# rvalue: correlation coefficient
# pvalue: two-sided p-value for a hypothesis whose null hypothesis is that the slope is zero, using Wald test with t-distribution of the test statistic.
# stderr: Standard error of the estimated gradient.

reg_line = np.poly1d([slope, intercept])  # function for regression line

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
fig.suptitle('Accuracy vs. Mean RT for 3 seconds condition')

ax1 = sns.scatterplot(x=rt, y=acc, ax=axes[0])
ax1.set(xlim=[0.5,2.5])

ax2 = sns.scatterplot(x=rt, y=acc, ax=axes[1])
regline = sns.lineplot(x=np.linspace(0.5,2.5), y=reg_line(np.linspace(0.5,2.5)), color='red')
ax2.set(xlim=[0.5, 2.5], title='corr-coeff r = {:.2}, p = {:.2}'.format(rvalue, pvalue))

fig.tight_layout()
fig.subplots_adjust(top=0.85)

fig.show()
````

**Answer**

The p-value p=0.043 < 0.05 indicates evidence against the null-hypothesis. We might reject the null-hypothesis
